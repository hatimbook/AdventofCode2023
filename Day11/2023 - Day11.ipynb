```python
import re
import math
import numpy as np
import itertools

with open('input.txt','r') as f:
    txt = f.read().strip().split('\n')
data = [[txt[r][c]  for c in range(len(txt[0]))] for r in range(len(txt))]
data = np.array([np.array(xi) for xi in data])
data[data == '.'] = 0
data[data == '#'] = 1
data = data.astype('int')

s = np.nonzero(data)

stars = []
for i in range(len(s[0])):
    stars.append([s[0][i], s[1][i]])
emptycols = np.sum(data, axis = 0)
emptyrows = np.sum(data, axis = 1)


for i in range(len(stars)):
    r, c = stars[i]
    stars[i][0] += ((np.nonzero(emptyrows[:r]==0))[0].size * 999999)
    stars[i][1] += ((np.nonzero(emptycols[:c]==0))[0].size * 999999)

starpairs = list(itertools.combinations(stars, 2))
print('Part2 = ', sum([compute_dist(x[0], x[1]) for x in starpairs])
```

```python
def compute_dist(a,b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])
```
