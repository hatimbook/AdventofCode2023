```python
with open('input.txt','r') as f:
    txt = f.read().strip()

txt = txt.split('\n')

num = ''
part_sum = 0
for r in range(len(txt)):
    row = txt[r]
    for c in range(len(row)):
        ch = row[c]
        if ch.isdigit():
            num = num + ch
        else:
            if num != '':
                neigh = get_neigh(r,c-len(num),len(num))
                if checkAllDots(neigh):
                    part_sum = part_sum + int(num)
                num = ''
    if num != '':
        neigh = get_neigh(r,c-len(num)+1,len(num)) #adding +1 to the column is critical since we are at end of row
        if checkAllDots(neigh):
                part_sum = part_sum + int(num)
        num = ''
print('Part1 = ', part_sum)
```

```python
def get_neigh(r,c,w):
    global txt
    n = []
    #get immediate neighbours from same row if they exist
    if c > 0:
        n.append(txt[r][c-1])
    if c+w < len(txt[0]):
        n.append(txt[r][c+w])
    #get neighbours from top and bottom rows if they exist
    for i in range(c-1, c+w+1):
        if i >= 0 and i < len(txt[0]):
            if r > 0:
                n.append(txt[r-1][i])
            if r+1 < len(txt):    
                n.append(txt[r+1][i])
    return n        
```

```python
def checkAllDots(n):
    for s in n:
        if s != '.':
            return True
    return False
```

```python
import math as m, re

board = list(open('input.txt'))
chars = {(r,c,board[r][c]): [] for r in range(140) for c in range(140) if board[r][c] not in '0123456789.'}

for r, row in enumerate(board):
    for n in re.finditer(r'\d+', row):
        edge = {(r,c, board[r][c]) for r in (r-1, r, r+1) for c in range(n.start()-1, n.end()+1) if r<140 and c<140}
        for o in edge & chars.keys():
            chars[o].append(int(n.group()))
p1 = sum(sum(p) for p in chars.values())
p2 = sum(m.prod(p) for p in chars.values() if len(p) == 2)

for key, values in chars.items():
    print(key,values)

print('Part 1 =',p1)
print('Part 2 =',p2)
```
