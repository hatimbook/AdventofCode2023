```python
def get_type(s):
    c = set(s)
    l = len(c)
    x = max([s.count(x) for x in c])
    if x==5:
        return 0
    elif x==4:
        return 1
    elif x==3:
        if l == 2: return 2
        else: return 3
    elif x==2:
        if l == 3: return 4
        else: return 5
    else:
        return 6
    
def get_type_with_Joker(s):
    c = set(s)
    l = len(c)
    x = max([s.count(x) for x in c])
    j = s.count('J')
    if (x==5) or (l==2 and j>=1):
        if j>5:
            print('Error!!')
        return 0
    elif (x==4) or (j==3 and l==3) or (j==2 and l==3) or (j==1 and x==3 and l==3):
        if j>3:
            print('Error!!')
        return 1
    elif (x==3 and l==2) or (l==3 and j==1):
        return 2
    elif (x==3 and l==3) or (x==2 and l==4 and j>=1):    
        return 3
    elif (x==2 and l== 3) or (x == 1 and l==4 and j>=1):
        if j > 0:
            print('Error!!')
        return 4
    elif (x==2 and l==4) or (l==5 and j>=1):    
        if j>1:
            print('Error!!')
        return 5
    else:
        if j>0:
            print("Error!!")
        return 6
```
